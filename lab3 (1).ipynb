# Arbol de decisión

Un árbol de decisión es un mapa de los posibles resultados de una serie de decisiones relacionadas. Permite que un individuo o una organización comparen posibles acciones entre sí según sus costos, probabilidades y beneficios. Se pueden usar para dirigir un intercambio de ideas informal o trazar un algoritmo que anticipe matemáticamente la mejor opción.

# K-Nearest Neighbors

Es un algoritmo basado en índices. Utiliza un método no paramétrico para la clasificación o regresión. Para los problemas de clasificación, el algoritmo consulta los k puntos que están más próximos al punto de muestra y devuelve la etiqueta que se utiliza con más frecuencia de su clase como la etiqueta predicha. Para los problemas de regresión, el algoritmo consulta los k puntos que están más próximos al punto de muestra y devuelve la media de sus valores característica como el valor predicho.

# LDA – Análisis de Discriminante Lineal.

El análisis discriminante lineal (LDA por sus siglas en inglés) es una clasificación de aprendizaje automático supervisado (binario o multimonial) y un método de reducción de dimensiones. LDA encuentra combinaciones lineales de variables que mejor “discriminan” las clases de la variable respuesta.

# Ensambles

a. Random Forest
Random Forest es un método de ensamble que construye múltiples árboles de decisión y los combina para mejorar la precisión y controlar el sobreajuste.


b. Gradient Boosting
Gradient Boosting es un método de ensamble que construye árboles de decisión secuencialmente, donde cada árbol corrige los errores del árbol anterior.

c. XGBoost
XGBoost es una implementación eficiente y optimizada de Gradient Boosting, con mejoras en la velocidad y el rendimiento.


d. AdaBoost
AdaBoost es un método de ensamble que ajusta secuencialmente modelos simples (como árboles de decisión débiles) a los datos, enfocándose en las observaciones mal clasificadas.
